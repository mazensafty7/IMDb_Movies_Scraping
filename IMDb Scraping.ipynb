# Import Libraries

```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
```

# Setup The Request

```python
base_url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
```

```python
driver = webdriver.Chrome()  # Ensure ChromeDriver is installed and accessible
driver.get(base_url)
```

```python
html = driver.page_source
```

# Make Soup Object

```python
soup = BeautifulSoup(html , "lxml")
```

# Scrap The Data

### Scrap Movies Titles

```python
main_page = soup.find_all("ul" , {"class" : "ipc-metadata-list ipc-metadata-list--dividers-between sc-a1e81754-0 iyTDQy compact-list-view ipc-metadata-list--base"})
```

```python
main_page
```

```python
titles = main_page[0].find_all("h3" , {"class" , "ipc-title__text"})
```

```python
len(titles)
```

```python
titles
```

```python
movies_titles = []
for title in titles:
    movies_titles.append(title.text.split(' ' , 1)[1].strip())
movies_titles
```

### Scrap Movies Year

```python
combined = main_page[0].find_all("span" , {"class" , "sc-6ade9358-7 exckou cli-title-metadata-item"})
```

```python
combined
```

```python
len(combined)
```

```python
year = []
for i in range(len(combined)):
    try:
        integer = int(combined[i].text)
        year.append(integer)
    except ValueError:
        pass
year
```

### Scrap Movies Duration

```python
duration = []
for i in range(len(combined)):
    if ("m" in combined[i].text) or ("h" in combined[i].text):
        duration.append(combined[i].text)
duration
```

### Scrap Movies MPAA

```python
mpaa_ratings = [
    "G",    
    "PG",   
    "PG-13",
    "R",    
    "NC-17",
    "Not Rated",
    "Approved",
    "Passed",
]
```

```python
MPAA = []
i = 2  # Start from index 2
while i < len(combined):
    if combined[i].text not in mpaa_ratings:
        MPAA.append("")
        i = i + 2
    else:
        MPAA.append(combined[i].text)
        i += 3
MPAA
```

### Scrap Movies Rating

```python
rating = main_page[0].find_all("span" , {"class" , "ipc-rating-star--rating"})
```

```python
rating
```

```python
len(rating)
```

```python
movies_ratings = []
for rate in rating:
    movies_ratings.append(rate.text)
movies_ratings
```

```python
len(movies_ratings)
```

# Make The DataFrame

```python
dic = {"Movie Name" : movies_titles , "Year" : year , "Duration" : duration , "MPAA" : MPAA , "Rate" : movies_ratings}
```

```python
dic
```

```python
df = pd.DataFrame(dic)
df.head()
```

```python
df.shape
```

```python
df.info()
```

# Data Cleaning

```python
df["Rate"] = df["Rate"].astype(float)
```

```python
df.info()
```
